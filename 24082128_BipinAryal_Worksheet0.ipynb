```python
#Task 1: Unit Conversion Program
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (str): Type of conversion (m_to_ft, ft_to_m, kg_to_lbs, lbs_to_kg, l_to_gal, gal_to_l)
    value (float): Value to be converted

    Returns:
    float: Converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type")


try:
    print("Choose Conversion Type:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    choice = int(input("Enter your choice: "))
    value = float(input("Enter value to convert: "))

    conversions = {
        1: "m_to_ft",
        2: "ft_to_m",
        3: "kg_to_lbs",
        4: "lbs_to_kg",
        5: "l_to_gal",
        6: "gal_to_l"
    }

    result = convert_units(conversions[choice], value)
    print("Converted Value:", result)

except ValueError:
    print("Invalid input. Please enter numeric values.")
except KeyError:
    print("Invalid conversion choice.")
```

```python
#Exercise on List Manipulation
def extract_every_other(lst):
    """Returns every other element starting from the first."""
    return lst[::2]


def get_sublist(lst, start, end):
    """Returns a sublist from start to end index inclusive."""
    return lst[start:end+1]


def reverse_list(lst):
    """Returns the reversed list using slicing."""
    return lst[::-1]


def remove_first_last(lst):
    """Removes first and last elements."""
    return lst[1:-1]


def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]


def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]


def reverse_skip(lst):
    """Returns every second element in reverse starting from second last."""
    return lst[-2::-2]


# Example usage
lst = [1,2,3,4,5,6]

print("Every other:", extract_every_other(lst))
print("Sublist:", get_sublist(lst,2,4))
print("Reverse:", reverse_list(lst))
print("Remove first last:", remove_first_last(lst))
print("First n:", get_first_n(lst,3))
print("Last n:", get_last_n(lst,2))
print("Reverse skip:", reverse_skip(lst))
```

```python
#Exercise on Nested Lists
def flatten(lst):
    """
    Flattens a nested list into a single list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


def access_nested_element(lst, indices):
    """
    Access element from nested list using indices.
    """
    for index in indices:
        lst = lst[index]
    return lst


def sum_nested(lst):
    """
    Returns sum of all elements in nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


def remove_element(lst, elem):
    """
    Removes all occurrences of element from nested list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


def find_max(lst):
    """
    Finds maximum element in nested list.
    """
    flat = flatten(lst)
    return max(flat)


def count_occurrences(lst, elem):
    """
    Counts occurrences of an element in nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


def deep_flatten(lst):
    """
    Deeply flattens nested lists.
    """
    return flatten(lst)


def average_nested(lst):
    """
    Calculates average of nested list elements.
    """
    flat = flatten(lst)
    return sum(flat) / len(flat)


# Example usage
nested = [[1,2],[3,[4,5]],6]

print("Flatten:", flatten(nested))
print("Access element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]],[1,2]))
print("Sum nested:", sum_nested(nested))
print("Remove element:", remove_element([[1,2],[3,2],[4,5]],2))
print("Max nested:", find_max(nested))
print("Count occurrences:", count_occurrences([[1,2],[2,3],[2,4]],2))
print("Deep flatten:", deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
print("Average nested:", average_nested([[1,2],[3,4],[5,6]]))
```
